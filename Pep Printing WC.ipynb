```python
stopwords = set(STOPWORDS)
```

```python
amazon_wc = WordCloud(
    background_color='lightgrey',
    max_words=2000,
    stopwords=stopwords)

amazon_wc.generate(amazon)
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 


plt.imshow(amazon_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
